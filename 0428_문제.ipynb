```python
import requests as req
res = req.get('http://ggoreb.com/quiz/special.html')
res.encoding = None
```

```python
html =res.text
idx = html.find('$$')
html = html[idx:]
html
```

```python
import re
pat = re.compile('[%$\^@_#()!&\[\]{}+*\\n]')
change = ''
result = pat.sub(change, html)
result
```

```python
import requests as req
res = req.get('http://ggoreb.com/quiz/pattern.html')
res.encoding = None
res.text
```

```python
html =res.text
idx = html.find('<!--')
html = html[idx:]
html
```

```python
import re 
pattern = re.compile('[a-z][A-Z]{3}([a-z]{1})[A-Z]{3}[a-z]')
result = pattern.findall(html)
print(result)
```
