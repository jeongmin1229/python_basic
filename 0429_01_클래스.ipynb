### 클래스
- 객체지향의 가장 기본적 개념
- 관련된 속성(모습)과 동작(기능)을 하나의 범주로 묶어 실세계의 사물을 표현
- 모델링 : 사물을 분석하여 필요한 속성과 동작 추출
- 캡슐화 : 모델 결과를 클래스로 포장(표현)
- 멤버 : 클래스를 구성하는 변수와 함수
- 메서드 : 클래스에 소속된 함수
- 클래스의 변수를 활용하려면 반드시 self 선언
- 메서드 자체의 코드만으로 동작할 때는 self 인수의 정보가 없어도 상관없음


```python
class Calculator:
    result = 0
    def adder(self, num):
        self.result += num
        return self.result
cal1 = Calculator()
cal2 = Calculator() 
result1 = cal1.adder(3)
result1 = cal1.adder(10) 
result2 = cal2.adder(13) 
result2 = cal2.adder(100) 

print(result1)
print(result2)
```

### ■ 클래스
- 클래스(Class) : 설계도면
- 객체(Object) : 클래스(설계도면)에 의해 만들어진 완성물(제품)
- 인스턴스(Instance) : 객체와 같은 의미

```python
class Computer:
    power = False
    def powerOn(self):
        self.power = True
        print('전원 ON')
    def powerOff(self):
        self.power = False
        print('전원 OFF')
        
com = Computer()
com.powerOn()
com.powerOff()
```

```python
class Letter:
    letter = ''
    
    def write(self, word):
        self.letter += word
    def show(self):
        return(self.letter)

letter1 = Letter()
letter1.write('가나다라')
letter1.write('카타파하')

letter2 = Letter()
letter2.write('ABCD')
letter2.write('WXYZ')
```

```python
print(letter1.show())
print(letter2.show())
print(Letter.letter)
```

```python
class Letter:
    letter = ''
    
    def write(self, word):
        self.letter += word
        Letter.letter += word # 이렇게 해야지 클래스 변수로 적용이 된다.
    def show(self):
        return(self.letter)

letter1 = Letter()
letter1.write('가나다라')
letter1.write('카타파하')

letter2 = Letter()
letter2.write('ABCD')
letter2.write('WXYZ')
```

```python
print(letter1.show())
print(letter2.show())
print(Letter.letter)
```

```python
class Letter:
    letter = 'asd'
    
    def __init__(self, data):
        self.data = data # 초기화를 해주면 딱히 변수선언을 하지 않아도 된다.
        
    def show(self):
        return(self.data)

letter1 = Letter('가나다라')
letter2 = Letter('ABCD')
```

```python
print(letter1.show())
print(letter2.show())
print(Letter.letter)
```
