```python
try:
    print(1)
    print(10)
except IndexError as e:
    print(e)
else:
    print('성공')
```

```python
try:
    print(1)
    print(10)
    lst = []
    lst[5]
except IndexError as e:
    print(e)
else:
    print('성공')
finally:
    print('무조건 실행')
```
