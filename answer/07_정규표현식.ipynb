```python
# 요구사항 : 뒷 7자리 숫자 중 첫번째 숫자는
#            1 2 3 4
import re
text = '791111-4234567'
# regex = '[0-9]{6}-\d{7}'
regex = '[0-9]{6}-[1-4]{1}\d{6}'
pattern = re.compile(regex)
res = pattern.match(text)
print(res)
```

```python
import re
text = '계좌번호는 123-12-123456'
# regex = '\d{3}-\d{2}-\d{6}'
regex = '1\d{2}-1\d{1}-\d{6}'
pattern = re.compile(regex)
res = pattern.match(text)
print(res)
```

```python
import re
text = '안녕하세요. 저는 OOO에서 태어났고... 열심히 하겠습니다. 감사합니다.'
regex = '^안녕하세요.*열심히.*감사합니다[.]$'
pattern = re.compile(regex)
res = pattern.match(text)
print(res)
```

```python
import re
text = '이름:       꼬렙,    전화번호:  1234'
regex = '.*이름:\s*([ㄱ-힣]+),\s*전화번호:\s*(\d+)'
pattern = re.compile(regex)
res = pattern.findall(text)
res
# print(res.group(1))
# print(res.group(2))
```

```python
import re
text = '대표자 : 꼬렙 사업자등록번호 : 100-99-23242324 대표전화 : 051-1234-5678'
regex = '(\d{3}-\d{2}-\d{8})'
pattern = re.compile(regex)
res = pattern.search(text)
print(res)
print(res.group(1))
```

```python
import re
text = '대표자 : 꼬렙 사업자등록번호 : 100-99-23242324 대표전화 : 051-1234-5678'
regex = '(\d{2,3}-\d{4}-\d{4})'
pattern = re.compile(regex)
res = pattern.search(text)
print(res.group())
```

#### 숨어있는 김첨지 찾아내기 (김첨지 단어가 몇개인지?)
- 김첨지, 김 첨지, 김  첨    지, 김첨 지, ....

```python
import requests
res = requests.get(
    'http://ggoreb.com/python/file/운수좋은날.txt'
)
res.encoding = None
res.text
```

```python
res.text
```

#### 김첨지 찾아내기

```python
import re
pat = re.compile('김\s*첨\s*지')
# pat = re.compile('김첨지')
result = pat.findall(res.text)
len(result), result
```

#### 길동이들 찾아내기

```python
import re
pat = re.compile('[ㄱ-힣]{1}길동')
result = pat.findall(res.text)
len(result), result
```

```python
import re
text = '010-8478-8181'
regex = '(\d{3})-\d{4}-(\d{4})'
pattern = re.compile(regex)
change = '\g<1>-****-\g<2>'
res = pattern.sub(change, text)
print(res)
```
