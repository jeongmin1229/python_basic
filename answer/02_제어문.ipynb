## while

```python
age = 41
while age > 30:
    print(age)
    age = age - 1
```

```python
import random
while True:
    number = random.randint(1, 10)  # 1 ~ 10 숫자 랜덤 추출
    print(number)
    if number == 5: break
```

```python
num = 1
total = 0

while num <= 1000:
    total += num  # total = total + num
    num = num + 1  # 증감식 (증가 / 감소)
    print(num, total)

print(total)
```

```python
num = 1
total = 0

while num <= 1000:
    num = num + 1  # 증감식 (증가 / 감소)
    total += num  # total = total + num
    print(num, total)

print(total)
```
