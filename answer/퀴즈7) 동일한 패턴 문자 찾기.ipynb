```python
import requests
res = requests.get(
    'http://ggoreb.com/quiz/pattern.html'
)
res.encoding = None
res.text
```

```python
html = res.text
idx = html.find('kAe')
html = html[idx:]
html
```

```python
import re
pat = re.compile('[^A-Z][A-Z]{3}([a-z]{1})[A-Z]{3}[^A-Z]')
result = pat.findall(html)
print(result)
print(''.join(result))
```
