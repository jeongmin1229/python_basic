#### 1. 04_seoul_기온.csv 파일 불러오기

```python
f = open('04_seoul_기온.csv', 'rt')
lines = f.readlines()
f.close()
```

```python
lines[:5]
```

```python
lines = lines[1:]
```

#### 2. 날짜와 최저기온 데이터 정리

```python
date_data = []
min_data = []
for line in lines:
    date_data.append( line.split(',')[0] )
    min_data.append( line.split(',')[2] )
```

#### 3. 시각화 준비 (환경설정)

```python
# matplotlib.pyplot 모듈 가져오기
import matplotlib.pyplot as plt
```
