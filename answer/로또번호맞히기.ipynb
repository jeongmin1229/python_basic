```python
import random

lotto = [3, 11, 12, 33, 45]  # 1등 번호
num = []
count = 0
while True:
  num.clear()  # 1등 번호를 다시 생성
  count += 1
  
  if count % 10000 == 0: print(count)
  
  while True:
    ran = random.randint(1, 45)
    if ran not in num:
      num.append(ran)
      if len(num) >= len(lotto): break
  num.sort()
  if lotto == num: break
print(count)
```
